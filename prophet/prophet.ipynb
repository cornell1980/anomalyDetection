```python
import pandas as pd
from datetime import datetime, timedelta
import yfinance as yf
from prophet import Prophet
```

```python
date_string = "2024-04-26"
DAYS_BACK = 120
end_date = datetime.fromisoformat(date_string)
start_date = (end_date - timedelta(days=DAYS_BACK)).strftime('%Y-%m-%d')
df = yf.download("IBM", start=start_date, end=end_date)
df.plot(y="Close")
```

```python
alpha = 0.98
model = Prophet(interval_width = alpha,
                yearly_seasonality = False,
                weekly_seasonality = False,
                changepoint_prior_scale=0.5)

model.fit(df)
```
